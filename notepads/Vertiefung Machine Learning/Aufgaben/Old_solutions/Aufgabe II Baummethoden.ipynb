# Imports

```python
from sklearn.tree import DecisionTreeRegressor
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

# Data Loading

```python
data = load_diabetes()
x = pd.DataFrame(data.data,columns=data.feature_names)
y = data.target
full = x.copy()
full["target"] = y
```

# Splitting

```python
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

# Training

```python
tree = DecisionTreeRegressor(max_depth=3)
tree.fit(x_train["bmi"].to_numpy().reshape(-1, 1),y_train)
```

# Visualisierung

```python
minimum, maximum = full["bmi"].min(),full["bmi"].max()
linspace = np.linspace(minimum,maximum)
plt.scatter(full["bmi"],full["target"],alpha=0.3)
plt.plot(linspace,tree.predict(linspace.reshape(-1, 1)))
```

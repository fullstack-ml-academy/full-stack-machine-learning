# Random Forest

## Imports

```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

## Datengrundlage

```python
data = pd.read_csv(os.path.join('..', '..', '..', '..', 'data', 'mpg.csv'))
x_train, x_test, y_train, y_test = train_test_split(data.drop(["mpg"],axis=1), data["mpg"], test_size=0.33, random_state=42)
```

## Einzelnen Decision Tree Regressor trainieren

```python
dt = DecisionTreeRegressor(max_depth=5)
dt.fit(x_train,y_train)
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(len(y_test)),y_test.sort_values(),c="#1ACC94",label="Realer Verbrauch")
plt.plot(range(len(y_test)),dt.predict(x_test.loc[y_test.sort_values().index]),c="#CC1B58",label="Decision Tree Vorhersage")
plt.title("Realer Benzinverbrauch und Vorhersage")
plt.xlabel("Geordnete Fahrzeuge nach Verbrauch")
plt.ylabel("Miles-per-Gallon")
plt.legend()
plt.show()
```

## Random Forest trainieren

```python
rf = RandomForestRegressor(max_depth=5)
rf.fit(x_train,y_train)
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(len(y_test)),y_test.sort_values(),c="#1ACC94",label="Realer Verbrauch")
plt.plot(range(len(y_test)),dt.predict(x_test.loc[y_test.sort_values().index]),c="#CC1B58",label="Decision Tree Vorhersage")
plt.plot(range(len(y_test)),rf.predict(x_test.loc[y_test.sort_values().index]),c="#7902E5",label="Random Forest Vorhersage")
plt.title("Realer Benzinverbrauch und Vorhersage")
plt.xlabel("Geordnete Fahrzeuge nach Verbrauch")
plt.ylabel("Miles-per-Gallon")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier 
from sklearn import tree
import numpy as np
```

```python
# Regression 1D

x = np.array([84,1323,282,957,1386,810,396,474,501,660,1260,1005,1110,1290]).reshape(-1,1)
y = np.array([44,97,30,51,95,51,44,41,21,40,90,83,61,92])

clf = DecisionTreeRegressor(max_depth=2, min_samples_split=7, random_state=1)
clf.fit(x, y)

fig = plt.figure(figsize=(20,20))
_ = tree.plot_tree(clf, 
                   feature_names=['studying'],  
                   filled=True)
```

```python
# Regression 2D

x = np.array([[84, 2],[1323, 4],[282, 3],[957, 3],[1386, 4],[810, 4],[396, 2],[474, 3],[501, 1],[660, 3],[1260, 4],[1005, 4],[1110, 2],[1290, 3]])
y = np.array([44,97,30,51,95,51,44,41,21,40,90,83,61,92])

clf = DecisionTreeRegressor(max_depth=3, min_samples_split=4, random_state=1)
clf.fit(x, y)

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=['studying', 'Semester'],  
                   filled=True)
```

```python
# Klassifikation 1D

x = np.array([84,1323,282,957,1386,810,396,474,501,660,1260,1005,1110,1290]).reshape(-1,1)
y = np.array([44,97,30,51,95,51,44,41,21,40,90,83,61,92])
y_binary = np.array([1 if i >= 50 else 0 for i in y])

clf = DecisionTreeClassifier(criterion='entropy', max_depth=3, min_samples_split=3, random_state=1)
clf.fit(x, y_binary)

fig = plt.figure(figsize=(10,10))
_ = tree.plot_tree(clf, 
                   feature_names=['studying'],  
                   class_names=['nicht bestanden', 'bestanden'],
                   filled=True)
```

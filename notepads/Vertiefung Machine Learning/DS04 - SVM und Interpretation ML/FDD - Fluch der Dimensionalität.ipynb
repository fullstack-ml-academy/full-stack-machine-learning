```python
# Angenommen es gibt 1.000 Datenpunkte, und es kommen neue Dimensionen hinzu mit zufälligen Werten für die 
# neuen Attribute, steigt die durchschnittliche Distanz
```

```python
def euclidean_distance(p,q):
    return np.sqrt(np.sum((q-p)**2))

basic_dataset = []
for x in np.arange(0,10):
    for y in np.arange(0,10):
        for z in np.arange(0,10):
            basic_dataset.append([x,y,z])

average_distances = []
for i in range(1,1000):
    larger_dataset = np.concatenate([basic_dataset,np.random.randint(0,10,size=(1000,i))],axis=1)
    #random_point = np.random.randint(0,10,size=(10*(i+3)))
    random_point = np.random.randint(0,10,size=((i+3)))
    average_distance = []
    for i in larger_dataset:
        average_distance.append(euclidean_distance(i,random_point))
    average_distances.append(np.mean(average_distance))
    
plt.plot(average_distances)
```

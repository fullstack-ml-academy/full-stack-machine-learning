```python
import numpy as np
import matplotlib.pyplot as plt
```

#

```python
np.linspace(0,10,11)
```

```python
np.arange(0,10)
```

```python
fig, axes = plt.subplots(1,2,figsize=(15,6))

axes[0].scatter(np.arange(0,10),np.ones(10),c="#1ACC94")
axes[0].set_title("Bei einer Dimension braucht es 10 Datenpunkte,\n um den Raum zu f√ºllen")
axes[0].set_xlabel("$x_1$")
axes[0].set_yticks(np.ones(10)," "*10)

x2d,y2d = np.meshgrid(np.arange(0,10),np.arange(0,10))
axes[1].scatter(x2d,y2d,c="#1ACC94")
axes[1].set_title("Bei zwei Dimensionen braucht es 100 Datenpunkte,\n um den Raum zu f√ºllen")
axes[1].set_xlabel("$x_1$")
axes[1].set_ylabel("$x_2$")
plt.show()
```

```python
x3d,y3d,z3d = np.meshgrid(np.arange(0,10),np.arange(0,10),np.arange(0,10))
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(projection='3d')
ax.scatter(x3d,y3d,z3d,c="#1ACC94")
ax.set_title("Bei drei Dimensionen braucht es 1.000 Datenpunkte,\n um den Raum zu f√ºllen")
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")
ax.set_zlabel("$x_3$")
plt.show()
plt.show()
```

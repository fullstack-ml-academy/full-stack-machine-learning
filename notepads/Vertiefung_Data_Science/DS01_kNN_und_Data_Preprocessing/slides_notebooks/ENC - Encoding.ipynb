# Encoding

## Imports

```python
import pandas as pd
import os
from sklearn.preprocessing import OneHotEncoder
```

## Datengrundlage

```python
data = pd.read_csv(os.path.join('..', '..', '..', '..', 'data', 'census.csv'))
```

```python
x = data[["education","marital-status"]]
y = data["target"]
```

## Encoding des Targets
Targetwerte können problemlos als Ziffern encoded werden

```python
y = y.str.strip().map({"<=50K":0,">50K":1})
```

## Encoding der X Werte

```python
encoder = OneHotEncoder(drop="first")
encoder.fit(x)
x_encoded = encoder.transform(x)
pd.DataFrame(x_encoded.todense(),columns=encoder.get_feature_names_out())
```

# t-Distributed Stochastic Neighbor Embedding (t-SNE)

## Imports

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torchvision.datasets as datasets
```

## Datengrundlage

```python
n = 5000
mnist = datasets.MNIST(root='./data', train=True, download=True, transform=None)
x = mnist.data.numpy().reshape(60000,-1)[:n]
y = mnist.targets.numpy()[:n]
```

## t-SNE trainieren

```python
tsne = TSNE(n_iter=1000)
embedding = tsne.fit_transform(x)
```

```python
plt.figure(figsize=(14,6))
for i in range(10):
    plt.scatter(embedding[np.where(y==i),:1],embedding[np.where(y==i),1:],label=str(i),alpha=0.4)
plt.legend()
plt.title("MNIST t-SNE Embedding")
plt.show()
```

## Vergleich mit PCA

```python
pca = PCA(n_components=2)
pca_embedding = pca.fit_transform(x)
```

```python
plt.figure(figsize=(14,6))
for i in range(10):
    plt.scatter(pca_embedding[np.where(y==i),:1],pca_embedding[np.where(y==i),1:],label=str(i),alpha=0.4)
plt.legend()
plt.title("MNIST PCA Embedding")
plt.show()
```

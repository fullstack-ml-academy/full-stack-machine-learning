# Deskriptive Statistik

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Datengrundlage

```python
x = np.array([84,1323,282,957,1386,810,396,474,501,660,1260,1005,1110,1290])
y = np.array([67,97,30,51,57,51,44,41,21,40,90,83,61,92])
```

## Grafische Darstellung

```python
plt.scatter(x,y)
plt.xlabel("Vorbereitungszeit auf Klausur in Minuten")
plt.ylabel("Klausurergebnis")
plt.title("Visualisierung der Datengrundlage")
plt.show()
```

## Arithmetisches Mittel

```python
mean = np.mean(y)
print("Das Arithmetisches Mittel (der Durchschnit) des Klausurergebnisses liegt bei {:.2f}".format(mean))
```

### 1. Lesen Sie den Datensatz mit Pandas ein. 

```python
# definieren der Imports am Anfang des Notebooks

import pandas as pd
import os
```

```python
# navigieren zum Dateipfad in dem die Daten liegen
file_path = os.path.join("..", "..", "..", "data", "census.csv")

# csv Datei mit Pandas einlesen
census = pd.read_csv(file_path)

# ausgeben der Spaltennamen im DataFrame
print(census.columns)

# die ersten 10 Zeilen des DataFrames anschauen
census.head(10)
```

```python
# Spalten bestimmen, die rausgenommen werden sollen
to_pop = ["occupation", "relationship", "capital-gain", "capital-loss", "native-country", "education", "workclass", "marital-status"]

# geht durch die Liste der Spalten und entfernt sie
for col in to_pop:
    census.pop(col)

# anzeigen der ersten 5 Zeilen
census.head()
```

```python
# kategorische Werte durch numerische Werte ersetzen

# definiere ein Dictionary um die Werte zu ersetzen
sex_dict = {" Male": 1, " Female": 0}

# ersetzen der Werte in der Spalte durch das Dictionary
census["sex"] = census["sex"].replace(sex_dict)

# dasselbe f√ºr die anderen kategorischen Spalten
race_dict = {
    " White": 3,
    " Black": 2,
    " Asian-Pac-Islander": 4,
    " Amer-Indian-Eskimo": 1,
    " Other": 0,
}
census["race"] = census["race"].replace(race_dict)

# das Dictionary kann auch direkt in der Funktion √ºbergeben werden
census["target"] = census["target"].replace({' >50K': 1, ' <=50K': 0})

census.head()
```

```python
# √úbersicht der statistischen Zusammenfassung einfach mit einem Aufruf m√∂glich
census.describe()
```

### 3. Teilen Sie die Daten mittels Train-Test-Split von Sklearn in Trainings- und Testdaten

```python
from sklearn.model_selection import train_test_split

# zuf√§lliges Teilen des DataFrames in train/test Datens√§tze basierend auf dem Anteil der Testdaten
train, test = train_test_split(census, test_size=0.3)
```

### 4. Analysieren Sie mit Numpy, ob die Features in Test- und Trainingsdatensatz sich in den Mittelwerten unterscheiden.

```python
import numpy as np

# definiere eine Funktion zum Berechnen des Mittelwerts einer Spalte
def calculate_mean(dataframe, feature):
    """Calculate the mean for a feature in a data frame using numpy."""
    return np.mean(dataframe[feature].to_numpy())


print('Train: {}'.format(train.shape))
print('Test: {}'.format(test.shape))

# Gehe durch die Feature um den Mittelwert f√ºr Trainings- und Testdaten zu bekommen
for feature in census.columns:
    print("Feature: {}, mean train: {}, mean test: {}".format(feature, calculate_mean(train, feature), calculate_mean(test, feature)))
```

```python
# Zus√§tzlich: Funktion um die Mittelwerte bei verschiedenen Aufteilungen zu vergleichen

def train_test_mean(split_rate: float):
    """Show the mean for all features and train/test data based on split rate.

    Args:
        split_rate (float): Number between 0 and 1 to indicate percentage of data to use for test.
    """

    train, test = train_test_split(census, test_size=split_rate)

    # Output information
    print("\n---------split rate: {}---------".format(split_rate))

    print("Train: {}".format(train.shape))
    print("Test: {}".format(test.shape))

    for feature in census.columns:
        print(
            "Feature: {}, mean train: {}, mean test: {}".format(
                feature, calculate_mean(train, feature), calculate_mean(test, feature)
            )
        )
```

```python
# Rufe die Funktion f√ºr verschieden Aufteilungen auf

train_test_mean(0.2)
train_test_mean(0.3)
train_test_mean(0.4)
```

### Unterschied arithmetischer Mittel, getrimmter Mittel und Median

```python
import numpy as np
from scipy import stats

a = np.array([1, 1, 1, 4, 13])
print(np.mean(a))
print(stats.trim_mean(a, 0.25))
print(np.median(a))
```

### Gewichtetes Mittel

```python
price = np.array([1.65, 1.66, 1.6, 1.69, 1.64])  # Durchschn. Spritprice
mean_price = np.mean(price)
print(round(mean_price, 3))
n = np.array([15, 10, 15, 40, 20])  # Anzahl der Tankstellen
w = n/np.sum(n)  # Gewichte
print(w)
print(sum(w))
weighted_mean = np.dot(price, w)
print(round(weighted_mean, 3))
```

### Datenverteilungen

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame.from_dict({"Bayern":[29296,13140],"NRW":[37937,17926],"Niedersachsen":[15469,8003],"Baden-Württemberg":[24548,11103],"Hessen":[14243,6293],"Sachsen":[6869,4057],"Brandenburg":[4211,2531],"Rheinland-Pfalz":[10604,4098],"Sachsen-Anhalt":[3762,2181],"Schleswig-Holstein":[5927,2911],"Thüringen":[3770,2120],"Mecklenburg-Vorpommern":[3338,1611],"Berlin":[9815,3664],"Saarland":[2782,984],"Hamburg":[4517,1852],"Bremen":[1583,680]},orient="index",columns=["Restaurants","Einwohner in T"])
data["Restaurants pro T Einwohner"] = data["Restaurants"]/data["Einwohner in T"]
data.sort_values("Restaurants")

plt.figure(figsize=(8,6))
plt.title("Boxplot")
plt.boxplot(data["Restaurants pro T Einwohner"], labels=["Restaurants pro T Einwohner"], medianprops=dict(color="#1ACC94"))
plt.xticks(rotation=45)
plt.show()
```

### Zusammenhangsmaße

```python
import random

a = np.array(range(100))
b = a * 2
c = a * (-0.1)
d = np.zeros(100) + np.random.randint(100, size=(len(a)))

fig, axs = plt.subplots(1, 3, figsize=(20,10))
axs[0].scatter(a, b)
axs[1].scatter(a, c)
axs[2].scatter(a, d)

plt.show()
```

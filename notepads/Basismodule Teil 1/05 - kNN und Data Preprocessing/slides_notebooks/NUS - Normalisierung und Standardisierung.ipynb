# Normalisierung und Standardisierung

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

## Datengrundlage

```python
data = np.random.normal(7000, 3000, size=(400, 2))
```

## Visualisierung Ausgangsdaten

```python
plt.figure(figsize=(8, 8))
plt.scatter(data[:, :1],data[:, 1:], c="#1ACC94")
plt.title("Ausgangsdaten")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.show()
```

## Normalisierung

```python
data_normalized = (data - np.min(data, axis=0)) / (np.max(data, axis=0) - np.min(data, axis=0))

# Vergleich mit sklearn Normalisierung
minmax = MinMaxScaler()
# Händische Lösung ist äquivalent zur sklearn Normalisierung
print("Abweichung zur sklearn Normalisierung: {:.4f}".format(np.sum(minmax.fit_transform(data) - data_normalized)))
```

## Standardisierung

```python
data_standardized = (data - np.mean(data, axis=0)) / np.std(data, axis=0)

# Vergleich mit sklearn Normalisierung
standard = StandardScaler()
# Händische Lösung ist äquivalent zur sklearn Normalisierung
print("Abweichung zur sklearn Standardisierung: {:.4f}".format(np.sum(standard.fit_transform(data) - data_standardized)))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].scatter(data[:, :1],data[:, 1:], c="#1ACC94")
axes[0].set_title("Originale Daten")
axes[1].scatter(data_normalized[:, :1],data_normalized[:, 1:], c="#1ACC94")
axes[1].set_title("Normalisierte Daten")
axes[2].scatter(data_standardized[:, :1],data_standardized[:, 1:], c="#1ACC94")
axes[2].set_title("Standardisierte Daten")
plt.show()
```

## Lage- und Streuungsmaße

```python
print("Spannweite der normalisierten Daten: {}".format(np.max(data_normalized, axis=0)-np.min(data_normalized, axis=0)))
print("Spannweite der standardisierten Daten: {}\n".format(np.round(np.max(data_standardized, axis=0) - np.min(data_standardized, axis=0), 2)))

print("Arithmetisches Mittel der normalisierten Daten: {}".format(np.round(np.mean(data_normalized, axis=0), 2)))
print("Arithmetisches Mittel der standardisierten Daten: {}\n".format(np.round(np.mean(data_standardized, axis=0), 2)))

print("Standardabweichung der normalisierten Daten: {}".format(np.round(np.std(data_normalized, axis=0), 2)))
print("Standardabweichung der standardisierten Daten: {}".format(np.round(np.std(data_standardized, axis=0), 2)))
```

## Histogramme

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].hist(data[:, :1], bins=30, color="#1ACC94")
axes[0].set_title("Originale Daten ($x_1$)")
axes[1].hist(data_normalized[:, :1], bins=30, color="#1ACC94")
axes[1].set_title("Normalisierte Daten ($x_1$)")
axes[2].hist(data_standardized[:, :1], bins=30, color="#1ACC94")
axes[2].set_title("Standardisierte Daten ($x_1$)")
plt.show()
```

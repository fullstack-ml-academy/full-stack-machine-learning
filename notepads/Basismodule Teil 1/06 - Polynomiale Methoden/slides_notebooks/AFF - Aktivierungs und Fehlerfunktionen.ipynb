# Aktivierungs- und Fehlerfunktionen

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Datengrundlage

```python
vector = np.linspace(-10,10,10000)
```

## Aktivierungsfunktionen

### Heaviside-Funktion

```python
def step(x):
    return np.where(x<0,0,1)
```

```python
plt.figure(figsize=(8,6))
plt.plot(vector, step(vector),c="#1ACC94")
plt.xlim(-5,5)
plt.ylim(-1.5,2.5)
plt.vlines(0,-1.5,6,ls="--",alpha=0.4)
plt.hlines(0,-5,5,ls="--",alpha=0.4)
plt.title("Heaviside-Funktion (Schrittfunktion)")
plt.show()
```

### Sigmoidfunktion

```python
def sigmoid(x):
    return 1.0/(1 + np.exp(-x))
```

```python
plt.figure(figsize=(8,6))
plt.plot(vector, sigmoid(vector),c="#1ACC94")
plt.xlim(-5,5)
plt.ylim(-1.5,2.5)
plt.vlines(0,-1.5,6,ls="--",alpha=0.4)
plt.hlines(0,-5,5,ls="--",alpha=0.4)
plt.title("Sigmoidfunktion")
plt.show()
```

### Tangens Hyperbolicus Funktion

```python
def tanh(x):
    return 1-(2/(np.exp(2*x)+1))
```

```python
plt.figure(figsize=(8,6))
plt.plot(vector, tanh(vector),c="#CC1B58")
plt.xlim(-5,5)
plt.ylim(-1.5,2.5)
plt.vlines(0,-1.5,6,ls="--",alpha=0.4)
plt.hlines(0,-5,5,ls="--",alpha=0.4)
plt.title("Tangens Hyperbolicus Funktion")
plt.show()
```

## Rectified Linear Unit (ReLU)

```python
def relu(x):
    return np.where(x>0,x,0)
```

```python
plt.figure(figsize=(8,6))
plt.plot(vector, relu(vector),c="#090EE1")
plt.xlim(-5,5)
plt.ylim(-1.5,2.5)
plt.vlines(0,-1.5,6,ls="--",alpha=0.4)
plt.hlines(0,-5,5,ls="--",alpha=0.4)
plt.title("ReLu")
plt.show()
```

## Exponential Linear Unit (ELU)

```python
def elu(x):
    return np.where(x>0,x,1*(np.exp(x) -1))
```

```python
plt.figure(figsize=(8,6))
plt.plot(vector, elu(vector),c="#20232E")
plt.xlim(-5,5)
plt.ylim(-1.5,2.5)
plt.vlines(0,-1.5,6,ls="--",alpha=0.4)
plt.hlines(0,-5,5,ls="--",alpha=0.4)
plt.title("ELU")
plt.show()
```

## Softmax Aktivierung

```python
def softmax(v):
    e = np.exp(v)
    return e / e.sum()

np.random.uniform(size=5)
```

```python
plt.figure(figsize=(10,6))
plt.title("Softmax Aktivierung")
plt.bar(["Ausgabeneuron {}".format(i) for i in range(5)], softmax(np.random.normal(size=5)),color="#1ACC94")
plt.show()
```

## Fehlerfunktionen

```python
def logloss(y, pred):
    return -(y*(np.log(pred)) + (1-y)*np.log(1-pred))
```

```python
predictions = np.linspace(0.0001,0.9999)

plt.figure(figsize=(8,6))
plt.plot(predictions, logloss(np.ones(50),predictions),label="Fehler fr $y=1$",c="#1ACC94")
plt.plot(predictions, logloss(np.zeros(50),predictions),label="Fehler fr $y=0$",c="#CC1B58")
plt.xlabel("Prdiktion $\hat{y}$")
plt.ylabel("Fehler")
plt.title("Log Loss / Kreuzentropie")
plt.legend()
plt.show()
```

```python
x1 = 1
x2 = 3
w1,w2 = 0.3,0.3
w3,w4 = 0.6,0.6
w5 = 0.8
w6 = 0.9

def s(x):
    return 1/(1+np.exp(-x))
```

```python
n1 = x1*w2+x2*w3
n2 = x1*w2+x2*w4
n3 = s(n1)*w5+s(n2)*w6
y = s(n3)
```
